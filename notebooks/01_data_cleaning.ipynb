```python
import pandas as pd

def load_large_jsonl(file_path, chunk_size=100000):
    print(f"Loading {file_path} in chunks of {chunk_size}...")
    
    # Initialize the reader as an iterator
    reader = pd.read_json(
        file_path, 
        lines=True, 
        chunksize=chunk_size, 
        dtype_backend="pyarrow"
    )
    
    chunks = []
    for i, chunk in enumerate(reader):
        # Append the processed chunk to our list
        chunks.append(chunk)
        print(f"  Processed chunk {i+1} (Total rows so far: {(i+1)*chunk_size})")
    
    # Concatenate all cleaned chunks into a single DataFrame
    full_df = pd.concat(chunks, ignore_index=True)
    return full_df

# --- Execution ---

# Load Reviews
df_reviews = load_large_jsonl("data/video_games_reviews.jsonl")
print(f"Reviews loaded! Final Shape: {df_reviews.shape}")

# Load Metadata
df_meta = load_large_jsonl("data/video_games_meta.jsonl")
print(f"Metadata loaded! Final Shape: {df_meta.shape}")
```

```python
# 1. Isolate the 800 relevant products from the metadata first to keep it fast
df_meta_subset = df_meta[df_meta['parent_asin'].isin(relevant_products)].copy()

# 2. Extract the leaf category (the last element of the list)
df_meta_subset['leaf_category'] = df_meta_subset['categories'].apply(
    lambda x: x[-1] if len(x) > 0 else None
)

# 3. Consolidate "Gamepads & Standard Controllers" into "Controllers"
df_meta_subset['leaf_category'] = df_meta_subset['leaf_category'].replace({
    "Gamepads & Standard Controllers": "Controllers"
})

# 4. Define your "Core Six" targets
target_categories = [
    "Games", 
    "Headsets", 
    "Gaming Mice", 
    "Gaming Keyboards", 
    "Consoles", 
    "Controllers"
]

# 5. Filter metadata for only these 6 clean sub-domains
df_meta_final = df_meta_subset[df_meta_subset['leaf_category'].isin(target_categories)]

# 6. Extract the definitive list of final product IDs
final_product_ids = df_meta_final['parent_asin'].unique()

# 7. Cut your main reviews dataframe to match exactly these products
df_reviews_final = df_reviews[df_reviews['parent_asin'].isin(final_product_ids)]
```
